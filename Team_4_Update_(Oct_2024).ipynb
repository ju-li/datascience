# Clean Completion Date

```python
# Check if completion_date and primary_completion_date are of the format YYYY-MM-DD or YYYY-MM

import re

def check_date_format(date_str):
  """Checks if a date string is in YYYY-MM-DD or YYYY-MM format."""
  pattern = r'^\d{4}-(0[1-9]|1[0-2])(-(0[1-9]|[12]\d|3[01]))?$'
  return bool(re.match(pattern, str(date_str))) or date_str is None

filtered_trials['completion_date_is_valid'] = filtered_trials['completion_date'].apply(check_date_format)
filtered_trials['primary_completion_date_is_valid'] = filtered_trials['primary_completion_date'].apply(check_date_format)

# Display rows with invalid date formats
invalid_completion_date_rows = filtered_trials[~filtered_trials['completion_date_is_valid']]
invalid_primary_completion_date_rows = filtered_trials[~filtered_trials['primary_completion_date_is_valid']]

print("Invalid completion_date rows:")
print(invalid_completion_date_rows)

print("\nInvalid primary_completion_date rows:")
print(invalid_primary_completion_date_rows)
```

```python
from datetime import datetime

def convert_completion_date(date_str):
  """Converts completion_date to date. If the date string is YYYY-MM, assume YYYY-MM-01"""
  if date_str is None:
    return None
  try:
    return datetime.strptime(date_str, '%Y-%m-%d').date()
  except ValueError:
    try:
      return datetime.strptime(date_str + '-01', '%Y-%m-%d').date()
    except ValueError:
      return None


filtered_trials['completion_date_v2'] = filtered_trials['completion_date'].apply(convert_completion_date)
filtered_trials['primary_completion_date_v2'] = filtered_trials['primary_completion_date'].apply(convert_completion_date)
```

```python
# Drop if both completion_date_v2 and primary_completion_date_v2 are None, or if both completion_date_v2 and primary_completion_date_v2 are less than a year ago

filtered_trials = filtered_trials.dropna(subset=['completion_date_v2', 'primary_completion_date_v2'], how='all')

from datetime import datetime, timedelta

# Define a threshold date (one year ago)
one_year_ago = datetime.now() - timedelta(days=365)

# Convert date columns to datetime objects
filtered_trials['completion_date_v2'] = pd.to_datetime(filtered_trials['completion_date_v2'], errors='coerce')
filtered_trials['primary_completion_date_v2'] = pd.to_datetime(filtered_trials['primary_completion_date_v2'], errors='coerce')

# Filter rows based on the condition
filtered_trials = filtered_trials[
    ~(
        (filtered_trials['completion_date_v2'] > one_year_ago) &
        (filtered_trials['primary_completion_date_v2'] > one_year_ago)
    )
]

filtered_trials
```
